**Part A - English Corpus (Alice in the wonderland)**

1) Text Preprocessing

****

We made used the PlainTextCorpusReader from nltk.corpus in order to segment sentences. Afterwards we made a 80-20 train-test split with random allocation and created the according files. We chose not to use .lower() methods or removing punctuation, because we wanted that information to remain for the RNN to work with.


2) Subword Segmentation

Judging performance could only be done for the english corpus and we hope that our expectation can be transferred to bengali.
Creating subwords with a vocabulary size of the alphabet yields as expected words with every letter as subword.
Expected for a medium vocabulary size are the most common subwords of length 2-4. We chose 400 as size, since with 200 the granularity seemed too fine, e.g. "_and" "_A nd" still were different among some other inconsistencies. 600  already lead to longer subwords, e.g. "_seemed" and expected would be "_see m ed".
For a large vocabulary size it would be desirable to cut off only small parts and keep most of the word meaning intact for the major subword.
The lower bound of 1500 was working best in this direction. Higher sizes often were not able to subword endings like "-ly" or "-ing" which we found less useful.

**6 Analysis**

**Subwords**

Taking OOV words into account can be a delicate task for unknown languages. Using subword representation can certainly help and is expected to work better when the internal structure of the words is preserved. This way we can choose the algorithm and granularity for subword generation accordingly. One of the questions in the project for us was, to what extent we can transfer knowledge about good hyperparameters for a known language to an unknown one.
For english, arriving at an intuitive subword representation did not end in a satisfying result. Desired would be an outcome that like “_work ed” for “worked”, where the stem preserved and in the later process, when an augmented train corpus is available, a former OOV word like “works” could be found because is was created from “work” and “s” in the generated data.
Unfortunately we observed that the subword representation often did not match our criteria and was splitting words at unintuitive positions, regardless of the granularity. 
This is one of the points that may decrease the OOV rate for the english corpus when elaborated further. 

**Models**

Using the same hyperparameter combination from the english models for bengali also resulted in models with a lower perplexity. The performance gain differs though. 
We observed the following decreases in perplexity for the baseline model versus the tuned model:

On alphabetical level the models improved likewise, but for larger subword granularities differences occur.
From this we conclude that there seem to be characteristics to the languages and knowledge transfer is not working equally. Whether this is due to influence of granularity or hyperparameter choice or probably a combination of both, we can not tell without an extensive grid search.
